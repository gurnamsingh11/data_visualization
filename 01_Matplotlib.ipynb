### Bar Plots

```python
plants = ['spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10, 25, 5, 30, 21]
germinated = [74, 88, 56, 69, 59]
```

```python
plt.figure(figsize=(6, 6))
plt.bar(plants, germinated, color="orchid", width=0.5)
plt.title("Germination Rates")
plt.ylabel("Number of Seedlings")
```

```python
plt.figure(figsize=(6, 6))
plt.bar(plants, died, color="red")
plt.title("Failure Rates")
plt.ylabel("Number of Dead Seedlings")
```

```python
plt.figure(figsize=(6, 6))
plt.bar(plants, died)
plt.bar(plants, germinated, width=0.4)
```

```python
plt.figure(figsize=(6, 6))
plt.bar(plants, died)
plt.bar(plants, germinated, width=0.4, align="edge")
```

```python
plt.figure(figsize=(6, 6))
plt.bar(plants, died, bottom=germinated, label="Failed Seedlings")
plt.bar(plants, germinated, label="Germinated Seedlings")
plt.legend(shadow=True, frameon=True, facecolor="white")
plt.title("Greenhouse Seedlings")
```

```python
plt.figure()
plt.barh(plants, germinated)
```

```python
import pandas as pd
titanic = pd.read_csv("Data_Analysis/Data/titanic.csv")
```

```python
titanic.head()
```

```python
titanic["age"] = pd.to_numeric(titanic["age"], errors="coerce")
```

```python
titanic.age
```

```python
plt.hist(titanic["age"])
```

```python
first_class = titanic[titanic["pclass"]==1]["age"]
```

```python
second_class = titanic[titanic["pclass"]==2]["age"]
```

```python
third_class = titanic[titanic["pclass"]==3]["age"]
```

```python
plt.hist(first_class, label="1st Class", alpha=0.5, bins=20)
plt.hist(second_class, label="2nd Class", alpha=0.5, bins=20)
plt.hist(third_class, label="3rd Class", alpha=0.5, bins=20)
plt.legend()
```

```python
# Use this data to generate the plot!
resorts = ["Crested Butte", "Aspen", "Telluride", "Wolf Creek"]
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]

# 74b9ff dark
```

```python
plt.figure(figsize=(8,6))
plt.bar(resorts,snow_last_24,color="#74b9ff",label="Last 24Hrs") 
plt.bar(resorts,snow_next_24,bottom=snow_last_24,color="#7ed6df",label="Next 24Hrs")
plt.title("CO Snow Storm Report")
plt.ylabel("Inches")
plt.legend()
```
