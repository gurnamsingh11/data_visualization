### Closer Look at Bar Plots

```python
salaries = pd.read_csv("Data_Analysis/Data/Salaries.csv", low_memory=False)
```

```python
salaries.head()
```

```python
salaries["BasePay"] = pd.to_numeric(salaries["BasePay"], errors="coerce").fillna(0)
salaries["OvertimePay"] = pd.to_numeric(salaries["OvertimePay"], errors="coerce").fillna(0)
salaries["OtherPay"] = pd.to_numeric(salaries["OtherPay"], errors="coerce").fillna(0)
```

```python
salaries.head()
```

```python
df = salaries[["EmployeeName", "BasePay", "OvertimePay", "OtherPay"]]
df.head()
```

```python
df.set_index("EmployeeName", inplace=True)
```

```python
df.head(10).plot(kind="bar")
```

```python
df.head(10).plot(kind="bar", stacked=True)
```

```python
df["BasePay"].sort_values(ascending=False).head(5).plot(kind="bar")
```

```python
df["BasePay"].sort_values(ascending=False).head(5).plot.bar(color="red", title="Highest Paid SF Employees")
```

```python
df["BasePay"].sort_values(ascending=False).head(5).plot.barh(color="red", title="Highest Paid SF Employees")
```

```python
df.sort_values("OvertimePay", ascending=False).head().plot(kind="barh", stacked=True)
```
