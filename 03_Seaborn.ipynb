### Line Plots

```python
flights = sns.load_dataset("flights")
```

```python
flights
```

```python
sns.lineplot(data=flights, x="year", y="passengers")
```

```python
flights.groupby("year")["passengers"].mean().plot(kind="line")
```

```python
sns.lineplot(data=flights, x="year", y="passengers", estimator="sum")
```

```python
sns.lineplot(data=flights, x="year", y="passengers", estimator="max")
```

```python
# Automatically sortsd X axis 
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
```

```python
trips = sns.load_dataset("taxis", parse_dates=["pickup", "dropoff"])
```

```python
trips
```

```python
trips["hour"] = trips["pickup"].dt.hour
```

```python
trips["color"]
```

```python
sns.lineplot(data=trips, x="hour", y="fare")
```

```python
sns.lineplot(data=trips, x="hour", y="fare", hue="payment")
```

```python
sns.lineplot(data=trips, x="hour", y="fare", hue="payment", style="color")
```

```python
sns.lineplot(data=trips, x="hour", y="fare", hue="payment", style="color", errorbar=('ci', 12))
```

### relplot()

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", col="sex")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", hue="smoker", col="sex")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", hue="smoker", col="sex", row="time")
```
